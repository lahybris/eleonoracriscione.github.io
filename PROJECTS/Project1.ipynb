### Player Information

- **Player:** Player's name
- **Position:** Player's position (PG, SG, SF, PF, C)
- **Age:** Player's age
- **Team:** Player's Team(s) Short Form
- **Games Played (G):** Number of games played
- **Games Started (GS):** Number of games started
- **Minutes Played (MP):** Number of minutes played

### Field Goals

- **Average Field Goals Made (FG)** 
- **Average Field Goals Attempted (FGA)**
- **Field Goals Percentage (FG%):** Average field goals percentage {(FG/FGA)*100}

### 3-Pointers

- **Average 3-Pointers Made (3P)**
- **Average 3-Pointers Attempted (3PA)**
- **3-Pointers Percentage (3P%):** Average 3-pointers percentage {(3P/3PA)*100}

### 2-Pointers

- **Average 2-Pointers Made (2P)**
- **Average 2-Pointers Attempted (2PA)**
- **2-Pointers Percentage (2P%):** Average 2-pointers percentage {(2P/2PA)*100}

### Other Shooting Statistics

- **Effective Field Goal Percentage (eFG%)**

### Free Throws

- **Average Free Throws Made (FT)**
- **Average Free Throws Attempted (FTA)**
- **Free Throw Percentage (FT%):** Average free throw percentage {(FT/FTA)*100}

### Rebounds

- **Average Offensive Rebounds (ORB) per game**
- **Average Defensive Rebounds (DRB) per game**
- **Average Total Rebounds (TRB) per game**

### Assists and Steals

- **Average Assists (AST) per game**
- **Average Steals (STL) per game**

### Blocks and Turnovers

- **Average Blocks (BLK) per game**
- **Average Turnovers (TOV) per game**

### Personal Fouls and Points

- **Average Personal Fouls (PF) per game**
- **Average Points (PTS) per game**

### Yearly MVP Statistics

- **Year (Year)**
- **Points Won (Pts Won):** Points won by the player to win the MVP award
- **Points Max (Pts Max):** Maximum points a player could get that season
- **Share (Share):** Share of total points a player received that season {Pts Won / Pts Max} (Response Variable)

### Team Information

- **Team (Team):** Player's Team(s) Long Form
- **Team Wins (W):** Team's wins that season
- **Team Losses (L):** Team's losses that season
- **Win/Loss Percentage (W/L%):** Team's win/loss percentage that season
- **Games Behind (GB)**
- **Average Points Scored per Game (PS/G):**
- **Average Points Allowed per Game (PA/G):**
- **Average Points Scored per Game (PS/G):**
- **Simple Rating System (SRS)**

### Scoring ability analysis

In this section, I aimed to analyze the scoring ability of players based on the year and their positions. The key elements in this analysis include field goal percentage, free throw percentage, and overall points per game. Firstly the data are sorted by points per games then the year is selected.


```python
top_players_df = selected_df.sort_values(by='PTS', ascending=False).head(20)
top_players_df.head()
```

```python
top_players_2018_df = selected_df[selected_df['Year'] == 2018]
top_players_2018_df.sort_values(by='PTS', ascending=False).head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(top_players_2018_df['FT'], top_players_2018_df['FG'], s=100, color='blue', alpha=0.7)
plt.xlabel('Free Throws')
plt.ylabel('Field Goals')
plt.title('Scatter Plot for Top 10 Players with Best Accuracy')

for i, player in top_players_2018_df.iterrows():
    if player['FT'] > 6:
        plt.text(player['FT'], player['FG'], player['Player'], fontsize=8, ha='right', va='bottom')
plt.show()
```

The first graph depicts a scatter plot representing the relationship between free throws and field goals in 2018. This scatter plot illustrates which player, as seen from the labels, was the most accurate, regardless of their position. Particularly, we observe that James Harden was an outlier compared to others, having the highest free throws per field goals relationship. Others, such as Anthony Davis, had a higher percentage of field goals compared to free throws, one of the highest among the rest.

```python
top_10_players = top_players_2018_df.sort_values(by='PTS', ascending=False)[:10]

plt.figure(figsize=(10, 6))
plt.bar(top_10_players['Player'], top_10_players['PTS'], color='Green')
plt.xlabel('Player')
plt.ylabel('Point per Game')
plt.title('Top 10 Players and Their Point per Game')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
```

The graph displays an Instagram with the highest points per player, specifically representing the top ten players for the year 2018, regardless of their position. The top three are Jim Harden, Anthony Davis, and LeBron James.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pos', y='FG', data=top_players_2018_df)
plt.xlabel('Position')
plt.ylabel('Field Goal')
plt.title('Effect of Position on Effective Field Goal')
plt.show()
```

At this point, it was relevant to study how the position influenced the goal. For this reason, a box plot is presented in this graph, comparing field goals to position. The overall range is pretty linear for every position; however, some positions, such as PF, show a high level of outliers compared to the rest. This suggests that the position is not so relevant in this study.

```python
pg_players_df = top_players_2018_df[top_players_2018_df['Pos'] == 'PG'].sort_values(by='FT', ascending=False)
pg_players_df = pg_players_df.groupby('Player').sum().reset_index()

pg_players_df['FT%'] = pg_players_df['FT'] / pg_players_df['FTA']
pg_players_df['FG%'] = pg_players_df['FG'] / pg_players_df['FGA']

pg_players_df = pg_players_df.dropna()

pg_players_df.head()
```

```python
plt.scatter(pg_players_df['FT%'], pg_players_df['FG%'],s=100, color='green', alpha=0.7)
plt.xlabel('Free Throw Percentage (FT%)')
plt.ylabel('Field Goal Percentage (FG%)')
plt.title('Scatter Plot for Top 10 Point Guards with Best Accuracy')

for i, player in pg_players_df.iterrows():
    if player['FG%'] > 0.6:
          plt.text(player['FT%'], player['FG%'], f"{player['Player']} (PG)", fontsize=8, ha='left', va='bottom')

plt.show()
```

To study these elements further, I have classified players according to the PG position. Specifically, I calculated the free throw percentage as free throws divided by free throw attempts and also the field goal percentage calculated in the same way. After calculating these values, I plotted a scatter plot, and it was possible to understand that, according to the PG position, athletes with the best field goal percentage and free throw percentage were David Stockton and Monte Morris.

```python
plt.scatter(pg_players_df['eFG%'], pg_players_df['MP'],s=100, color='green', alpha=0.7)
plt.xlabel('Effective Field Goals')
plt.ylabel('Minutes Played per Game')
plt.title('Scatter Plot for PG ')

for i, player in pg_players_df.iterrows():
    if player['MP'] < 5:
          plt.text(player['eFG%'], player['MP'], f"{player['Player']} (PG)", fontsize=8, ha='left', va='bottom')

plt.show()
```

In the correlation analysis, a peculiar relationship emerged between effective field goals and minutes played per game. When plotting this relationship for the position of point guards, it becomes evident that David Stockton exhibits an unusual behavior. He is the player with the fewest minutes per game played but the highest effective field goal percentage.
