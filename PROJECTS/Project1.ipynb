### Player Information

- **Player:** Player's name
- **Position:** Player's position (PG, SG, SF, PF, C)
- **Age:** Player's age
- **Team:** Player's Team(s) Short Form
- **Games Played (G):** Number of games played
- **Games Started (GS):** Number of games started
- **Minutes Played (MP):** Number of minutes played

### Field Goals

- **Average Field Goals Made (FG)** 
- **Average Field Goals Attempted (FGA)**
- **Field Goals Percentage (FG%):** Average field goals percentage {(FG/FGA)*100}

### 3-Pointers

- **Average 3-Pointers Made (3P)**
- **Average 3-Pointers Attempted (3PA)**
- **3-Pointers Percentage (3P%):** Average 3-pointers percentage {(3P/3PA)*100}

### 2-Pointers

- **Average 2-Pointers Made (2P)**
- **Average 2-Pointers Attempted (2PA)**
- **2-Pointers Percentage (2P%):** Average 2-pointers percentage {(2P/2PA)*100}

### Other Shooting Statistics

- **Effective Field Goal Percentage (eFG%)**

### Free Throws

- **Average Free Throws Made (FT)**
- **Average Free Throws Attempted (FTA)**
- **Free Throw Percentage (FT%):** Average free throw percentage {(FT/FTA)*100}

### Rebounds

- **Average Offensive Rebounds (ORB) per game**
- **Average Defensive Rebounds (DRB) per game**
- **Average Total Rebounds (TRB) per game**

### Assists and Steals

- **Average Assists (AST) per game**
- **Average Steals (STL) per game**

### Blocks and Turnovers

- **Average Blocks (BLK) per game**
- **Average Turnovers (TOV) per game**

### Personal Fouls and Points

- **Average Personal Fouls (PF) per game**
- **Average Points (PTS) per game**

### Yearly MVP Statistics

- **Year (Year)**
- **Points Won (Pts Won):** Points won by the player to win the MVP award
- **Points Max (Pts Max):** Maximum points a player could get that season
- **Share (Share):** Share of total points a player received that season {Pts Won / Pts Max} (Response Variable)

### Team Information

- **Team (Team):** Player's Team(s) Long Form
- **Team Wins (W):** Team's wins that season
- **Team Losses (L):** Team's losses that season
- **Win/Loss Percentage (W/L%):** Team's win/loss percentage that season
- **Games Behind (GB)**
- **Average Points Scored per Game (PS/G):**
- **Average Points Allowed per Game (PA/G):**
- **Average Points Scored per Game (PS/G):**
- **Simple Rating System (SRS)**

Source : https://www.kaggle.com/datasets/dhruvsuryavanshi/nba-player-data-from-2003-to-2022/

# Objective: 
### To analyze the accuracy of the score, minutes played, and potential for accuracy for each player
---

```python
columns_to_extract = ['Player', 'PTS', 'MP', 'FG', 'FGA', 'FT', 'FTA', 'eFG%','G']
selected_df = stats[columns_to_extract].copy()
selected_df.head()
```

```python
selected_df.info()
```

```python
selected_df['Player'] = selected_df['Player'].astype('category')
```

3. Data Visualization

```python
f, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
sns.boxplot(data=selected_df['PTS'], ax=axes[0])
sns.boxplot(data=selected_df['eFG%'], ax=axes[1])
axes[0].set_title('Points scored')
axes[1].set_title('Effective field goal percentage')
plt.tight_layout()
plt.show()
```

```python
f, axes = plt.subplots(2, 2, figsize=(10, 4), sharex=True)
sns.distplot(selected_df['FG'], kde=True, ax=axes[0,0])
sns.distplot(selected_df['FGA'], kde=True, ax=axes[0,1])
sns.distplot(selected_df['FT'], kde=True, ax=axes[1,0])
sns.distplot(selected_df['FTA'], kde=True, ax=axes[1,1])
axes[0,0].set_title('Field Goals Made')
axes[0,1].set_title('Field Goals Attempted')
axes[1,0].set_title('Free throws made')
axes[1,1].set_title('Free throws attempted')
plt.tight_layout()
plt.show()
```

```python
sns.distplot(selected_df['eFG%'], kde=True)
plt.xlabel('Effective field goal percentage')
plt.show()
```

```python
selected_df['FG%']=selected_df['FG']/selected_df['FGA']
selected_df.head()
```

```python
sns.distplot(selected_df['FG%'], kde=True)
plt.xlabel('Effective field goal percentage')
plt.show()
```

```python
selected_df['FT%']= selected_df['FT']/selected_df['FTA']
selected_df.head()
```

```python
sns.distplot(selected_df['FT%'], kde=True)
plt.xlabel('Effective field goal percentage')
plt.show()
```

```python
selected_df['AvgMPPerGame'] = selected_df['MP'] / selected_df['G']
selected_df.head()
```
